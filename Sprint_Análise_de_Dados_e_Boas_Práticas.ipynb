## Descrição do Problema

A Starbucks é uma das maiores redes de cafeterias do mundo, com milhões de pedidos processados diariamente em diferentes canais e formatos de atendimento. Entender como os clientes se comportam nesse ambiente é uma questão relevante tanto do ponto de vista de negócios quanto de análise de dados.

Este trabalho analisa o dataset Starbucks Customer Ordering Patterns, disponível na plataforma Kaggle, que reúne informações sobre pedidos, canais de atendimento, perfil dos clientes e valores gastos. O objetivo é investigar quais fatores se relacionam com o valor dos pedidos e identificar padrões no comportamento de consumo.

Para orientar essa investigação, foram definidas três hipóteses, que serão apresentadas e respondidas ao longo do trabalho com base nos dados analisados.

## Hipóteses do Problema

As hipóteses definidas para esta análise são:

1. Clientes que realizam pedidos em determinados períodos do dia tendem a apresentar maior valor gasto.

2. O canal de atendimento influencia o valor dos pedidos realizados.

3. Clientes mais frequentes tendem a apresentar maior valor médio por pedido.

Essas hipóteses serão avaliadas ao longo da análise exploratória e respondidas com base nos padrões identificados nos dados.

## Tipo de Problema

esse trabalho tem como objetivo a realização de uma análise exploratória dos dados para identificar padrões de comportamento de compra e responder as hipoteses levantadas, não tem como objetivo a construção de um modelo preditivo.

**Dessa forma, o problema se caracteriza como não supervisionado, já que não há uma variável alvo a ser prevista, e o foco está na identificação das relações e padrões entre as variáveis disponíveis.**

## Atributos do Dataset
Antes de iniciar a análise, é importante conhecer a estrutura da base de dados. O dataset conta com 100.000 registros distribuídos em 20 atributos, que cobrem desde informações sobre o pedido em si até características do cliente e do canal de atendimento.

Os atributos presentes na base são:

- `customer_id`: identificador do cliente
- `order_id`: identificador único do pedido
- `order_date`: data em que o pedido foi realizado
- `order_time`: horário em que o pedido foi realizado
- `day_of_week`: dia da semana em que o pedido foi feito
- `order_channel`: canal de atendimento utilizado no pedido
- `store_id`: identificador da loja
- `store_location_type`: tipo de localização da loja
- `region`: região geográfica da loja
- `customer_age_group`: faixa etária do cliente
- `customer_gender`: gênero do cliente
- `is_rewards_member`: indica se o cliente participa do programa de recompensas
- `cart_size`: quantidade de itens no pedido
- `num_customizations`: número de customizações realizadas no pedido
- `total_spend`: valor total gasto no pedido
- `fulfillment_time_min`: tempo de preparação/entrega do pedido em minutos
- `drink_category`: categoria principal da bebida
- `has_food_item`: indica se o pedido inclui item de comida
- `order_ahead`: indica se o pedido foi feito com antecedência
- `customer_satisfaction`: nível de satisfação do cliente

### Distribuição dos Dados

Nesta etapa, as distribuições das principais variáveis do dataset são analisadas com o objetivo de entender o comportamento geral dos dados e identificar padrões que possam estar relacionados às hipóteses levantadas.

```python
# criação da variável de período do dia, manhã, tarde e noite
# essas variáveis também serão utilizadas na sessão Análise de Relações entre Variáveis
df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M')

def periodo_dia(hora):
    if hora.hour < 12:
        return 'Manhã'
    elif hora.hour < 18:
        return 'Tarde'
    else:
        return 'Noite'

df['periodo_dia'] = df['order_time'].apply(periodo_dia)
```

```python
# distribuição dos pedidos por período do dia
plt.figure(figsize=(7,5))
ax = sns.countplot(x='periodo_dia', data=df)

plt.title('Distribuição dos Pedidos por Período do Dia')
plt.xlabel('Período do Dia')
plt.ylabel('Quantidade')

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.show()
```

**Analisando o resultado: Período do Dia**

A distribuição dos pedidos por período do dia mostra que a maior parte das compras ocorrem durante o período da manhã, já o período da tarde vem em segundo lugar, enquanto o período da noite apresenta uma quantidade menor de pedidos.

Esse comportamento de consumo é mais intenso nas primeiras horas do dia, o que pode estar relacionado a hábitos como café da manhã e início das atividades diárias.

Essa análise representa apenas a quantidade de pedidos por período e não o valor gasto, essa relação será explorada nas etapas seguintes.
Foi criada a variável categórica `periodo_dia`, permitindo agrupar os pedidos em faixas para as análises seguintes.


```python
#distribuição dos canais de pedidos
plt.figure(figsize=(7,5))
ax = sns.countplot(x='order_channel', data=df)
plt.title('Distribuição dos Canais de Pedido')
plt.xlabel('Canal de Pedido')
plt.ylabel('Quantidade')

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.show()
```

**Analisando o resultado: Distribuição dos canais de pedidos**

A distribuição dos canais de pedido mostra que o volume de pedidos não é uniforme entre os diferentes meios de atendimento.

A maior concentração de pedidos no aplicativo mobile indica uma preferência dos clientes por esse canal. Por outro lado, o canal de kiosk apresenta a menor quantidade de pedidos.

Essa diferença na quantidade de pedidos por canal é relevante para as análises posteriores, especialmente na investigação sobre a influência do canal no valor dos pedidos realizados.

```python
#distribuição dos valores de pedidos e as frequencias
plt.figure(figsize=(7,5))
sns.histplot(df['total_spend'], bins=30, kde=True)

plt.title('Distribuição do Valor Total dos Pedidos')
plt.xlabel('Valor do Pedido')
plt.ylabel('Frequência')

plt.show()
```

**Analisando o resultado: Total gasto**

A distribuição do valor total dos pedidos mostra uma concentração entre 10 e 16, com pico em torno de 14, o que bate com a média e mediana encontradas na estatística descritiva.
A curva tem uma leve cauda à direita, indicando que pedidos com valores mais altos existem mas são menos frequentes.

Essa distribuição será importante nas análises seguintes, especialmente para avaliar se o canal de atendimento ou o período do dia conseguem explicar as variações observadas no valor dos pedidos.

```python
#distribuição da quantidade de itens por carrinhos
plt.figure(figsize=(7,5))
sns.histplot(df['cart_size'], bins=10, kde=False)

plt.title('Distribuição do Tamanho do Carrinho')
plt.xlabel('Quantidade de Itens')
plt.ylabel('Frequência')

plt.show()
```

**Analisando o resultado: Tamanho do Carrinho**

A distribuição do tamanho do carrinho mostra que a maioria dos pedidos tem entre 3 e 4 itens, com pico em 3, confirmando o que a estatística descritiva já havia indicado. Pedidos com mais itens existem mas aparecem com frequência bem menor, mostrando que compras mais volumosas são pontuais.

Essa informação é relevante para as análises seguintes, já que a quantidade de itens no carrinho tende a influenciar diretamente o valor total gasto, e entender essa relação vai ajudar a interpretar melhor os resultados das hipóteses.

### Análise de Relações entre Variáveis

Com a distribuição das variáveis já entendida, o próximo passo é analisar como elas se relacionam entre si. O objetivo aqui é identificar padrões que ajudem a avaliar as três hipóteses levantadas, entendendo se o período do dia, o canal de atendimento e a frequência de compra realmente têm alguma influência no valor dos pedidos.

```python
# Analise de valor de pedidos por canal
plt.figure(figsize=(8,5))
sns.boxplot(x='order_channel', y='total_spend', data=df)

plt.title('Valor dos Pedidos por Canal de Atendimento')
plt.xlabel('Canal de Pedido')
plt.ylabel('Valor do Pedido')

plt.show()
```

**Analisando o resultado: Valor dos pedidos por canal de atendimento**

O gráfico deixa bem claro que o Mobile App se destaca dos demais canais, com mediana visivelmente mais alta e uma distribuição concentrada em faixas de valor superiores. Os canais Drive-Thru, Kiosk e In-Store Cashier se comportam de forma semelhante entre si, com medianas próximas e valores bem mais baixos em comparação ao app.

Existem outliers em todos os canais, representando pedidos pontuais com valores acima do padrão, mas isso não altera o padrão geral observado.

Esses resultados são uma evidência importante para a hipótese 2, indicando que o canal de atendimento tem sim influência sobre o valor dos pedidos, com destaque para o Mobile App como o canal associado aos maiores valores de compra.


```python
# análise da relação entre período do dia e valor dos pedidos
# as variáveis de período do dia, manhã, tarde e noite já foram criadas na sessão distribuição de dados e serão também utilizadas nessa sessão
plt.figure(figsize=(8, 5))
sns.boxplot(x='periodo_dia', y='total_spend', data=df)

plt.title('Valor dos Pedidos por Período do Dia')
plt.xlabel('Período do Dia')
plt.ylabel('Valor do Pedido')

plt.show()
```

**Analisando o resultado: Valor dos pedidos por período do dia**

Diferente do que foi observado no canal de atendimento, o gráfico de período do dia mostra um comportamento bastante uniforme entre manhã, tarde e noite. As medianas e a dispersão dos valores são praticamente idênticas entre os três períodos, o que sugere que o horário do pedido não tem influência relevante sobre o valor gasto.

Existem outliers em todos os períodos, com pedidos pontuais de valor mais elevado, mas esses casos não alteram o padrão geral visto aqui.

Esses resultados já apontam para uma resposta possível negativa à hipótese 1, indicando que o período do dia não é um fator que explica as variações no valor dos pedidos.

```python
# Agrupamento por cliente, quantidade de pedidos e valor médio gasto
df_cliente = df.groupby('customer_id').agg({
    'order_id': 'count',
    'total_spend': 'mean'
}).reset_index()
```

```python
# Análise da relação entre frequência de pedidos e valor médio gasto
plt.figure(figsize=(8,5))
sns.scatterplot(x='order_id', y='total_spend', data=df_cliente)

plt.title('Relação entre Frequência de Pedidos e Ticket Médio')
plt.xlabel('Quantidade de Pedidos')
plt.ylabel('Ticket Médio')

plt.show()
```

**Analisando o resultado: Relação entre frequência de pedidos e ticket médio**

O gráfico mostra que o ticket médio dos clientes se mantém em uma faixa bastante parecida, independentemente de quantos pedidos cada um realizou.
Clientes com poucos pedidos apresentam maior variação nos valores, com alguns casos de ticket mais elevado, enquanto os clientes mais frequentes tendem a se concentrar em uma faixa mais estável.

O que chama atenção é que não há nenhuma tendência de aumento no ticket médio conforme a frequência de pedidos cresce, o que já direciona para uma resposta negativa à hipótese 3. Clientes mais frequentes necessáriamente não gastam mais em cada compra, simplesmente compram com mais regularidade.

### Matriz de Correlação

A matriz de correlação mostra a relação linear entre as variáveis numéricas do dataset, indicando a força e a direção dessas relações. Valores próximos de 1 indicam forte correlação positiva, próximos de -1 indicam forte correlação negativa, e próximos de 0 indicam ausência de relação linear significativa.

```python
# matriz de correlação das variáveis numéricas mais relevantes para análise
df['valor_por_item'] = df['total_spend'] / df['cart_size']

colunas_corr = [
    'total_spend',
    'cart_size',
    'num_customizations',
    'valor_por_item'
]

corr = df[colunas_corr].corr().round(2)
corr
```

```python
# matriz de correlação
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Matriz de Correlação')
plt.show()
```

**Analisando o resultado: Matriz de correlação**

A matriz de correlação ajuda a entender melhor o que está por trás das variações no valor dos pedidos observadas nas análises anteriores. O resultado mais expressivo é a correlação de 0,90 entre o valor total dos pedidos (total_spend) e o tamanho do carrinho (cart_size), confirmando que a quantidade de itens é o principal fator que explica as diferenças no valor gasto.

Esse achado é relevante para as hipóteses analisadas, pois sugere que fatores como período do dia e frequência de compra provavelmente não influenciam o valor dos pedidos justamente porque não alteram o tamanho do carrinho. Já o canal de atendimento, que se mostrou associado a maiores valores no Mobile App, pode estar relacionado a um perfil de pedido com mais itens ou mais customizações, o que explicaria a diferença observada.

## Pré-Processamento dos dados

Durante o processo desenvolvido, foram realizadas transformações pontuais nos dados com o objetivo de apoiar as análises das hipóteses.

Durante a análise exploratória, foram criadas duas variáveis derivadas que merecem ser destacadas como parte do pré-processamento realizado.

Não foi necessário aplicar técnicas de normalização ou padronização, uma vez que as análises realizadas são de caráter exploratório.

## One-hot Encoding:

Como parte do pré-processamento, foi aplicada a técnica de one-hot encoding nas variáveis categóricas mais relevantes para as hipóteses do trabalho, order_channel e periodo_dia, transformando-as em variáveis numéricas binárias.
